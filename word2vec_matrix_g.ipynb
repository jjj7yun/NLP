```python
df = pd.read_excel('./Topic_Modeling/Data/전처리찐완성.xlsx')
```

```python
text = [i.strip().split() for i in df['nouns'].tolist()]
```

```python
len(text)
```

```python
ntx=[]
for li in text:
    nli=[]
    for word in li:
        nword = word[:-1]
        nli.append(nword)
    ntx.append(nli)
```

```python
len(ntx)
```

```python
ntx[0][0]
```

```python

import gensim
from gensim.models import KeyedVectors

DIM=100          #차원수
WINDOW_SIZE=10   #학습되는 주변단어 수
EPOCH =200       #반복학습 횟수


#word2vec model create
model = gensim.models.Word2Vec(ntx, vector_size = DIM,
                            min_count=10, window = WINDOW_SIZE,  #min_count는 빈도기준 학습제외
                            sg=1,epochs = EPOCH)     #sg=1 은 skip_gram, sg=0은 bag-of-bow



model.save('1210noun')
```
