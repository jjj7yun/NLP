## 텍스트 전처리

```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.stem import WordNetLemmatizer
from nltk.tag import pos_tag

from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS as sklearn_stop_words
import re
```

```python
ceo=list(df['CEO Message'])
material=list(df['Materiality Assessment'])
```

```python
print(len(ceo))
print(len(material))
```

```python
pattern=re.compile(r'([\t\r()-.,:;!?`‘\'"�‧|%“])+')
```

```python
stop_words=nltk.corpus.stopwords.words('english')
stop_words=list(set(stop_words).union(set(sklearn_stop_words)))
tokenizer=TreebankWordTokenizer()
lemma = WordNetLemmatizer()
```

```python
remove_bracket=['[',']','step']
stop_words.extend(remove_bracket)
len(stop_words)
```

```python
def prepro(text):
    a=text.lower()
    b=re.sub(pattern,'',a)
    b=re.sub(r'\[\[(?:[^\]|]*\|)?([^\]|]*)\]\]', '', b)
    c=re.sub('\n',' ',b)
    tagged_list = pos_tag(tokenizer.tokenize(c))
    ad_and_nouns = [t[0] for t in tagged_list if (t[1] == "NN")or(t[1]=="NNP") or(t[1]=='JJ')]
    e_list= [lemma.lemmatize(i) for i in ad_and_nouns]
    result=[w for w in e_list if w not in stop_words]
#     for w in e_list: 
#         if w not in stop_words: 
#             result.append(w)
    return result
    
```

```python
ceo[0]
```

```python
prepro(ceo[0])
```

```python
ceo_list=[' '. join(prepro(i)) for i in ceo]
ceo_list[:2]
    
```

```python
material_list=[' '. join(prepro(j)) for j in material]
material_list[:2]
```

```python
df['after_ceo']=ceo_list
```

```python
df['after_material']=material_list
```
