```python
df=pd.DataFrame(columns=['filelist','rawdata'])
```

```python
directory = './Data/subtext/'
filelist = os.listdir(directory)
df['filelist']=filelist
```

```python
for i in range(len(df)):
    j = df['filelist'][i]
    filename = directory + j
    f=open(filename, 'r', encoding='utf-8')
    txt = f.read()
    f.close()
    df['rawdata'][i] = txt

df
```

```python
df['filein']=df.filelist

for i in range(len(df)):
    df.filein[i] = df.filein[i][:3]
    df.filein[i] = re.sub('[.]','',df.filein[i])
    df.filein[i] = re.sub('[A-z]','',df.filein[i])
df
```

```python
for i in range(len(df)):
    if len(df.filein[i]) == 1 :
        df.filein[i] = '00' + df.filein[i]
    if len(df.filein[i]) == 2 :
        df.filein[i] = '0' + df.filein[i]  
df =df.sort_values('filein')
df
```

```python
df= df.set_index('filein')
df= df.set_index('filelist')
df.head()
```

```python
sub=[]
tmp_num=[]
for i in df.rawdata : 
    sub.extend(i.split('@@@'))
    tmp_num.append(len(i.split('@@@')))
```

```python
print(len(sub))
print(len(tmp_num))
```

```python
tmp_num[:5]
```

```python
df['split_num']=tmp_num
df.head()
```

```python
filelist=df.index
nums=df['split_num']
```

```python

high_index=[]
low_index=[]
for i, k in zip(filelist, nums):
    for j in range(k):
        high_index.append(i)
        low_index.append(j)
```

```python
print(len(high_index))
print(len(low_index))
```
